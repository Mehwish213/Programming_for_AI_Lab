```python
from collections import deque

CAP_A = 4
CAP_B = 3

def is_goal(state):
    return state[0] == 2  

def get_next_states(state):
    a, b = state
    next_states = []

    next_states.append(((CAP_A, b), "Fill A"))

    next_states.append(((a, CAP_B), "Fill B"))

    next_states.append(((0, b), "Empty A"))

    next_states.append(((a, 0), "Empty B"))

    transfer = min(a, CAP_B - b)
    next_states.append(((a - transfer, b + transfer), "Pour A → B"))

    transfer = min(b, CAP_A - a)
    next_states.append(((a + transfer, b - transfer), "Pour B → A"))

    return next_states


def dfs():
    stack = [((0, 0), [])]   
    visited = set()

    while stack:
        (state, path) = stack.pop()

        if state in visited:
            continue

        visited.add(state)

        if is_goal(state):
            print("Goal Reached!\n")
            for step in path:
                print(step)
            print("Final State:", state)
            return

        for next_state, rule in get_next_states(state):
            if next_state not in visited:
                stack.append((next_state, path + [f"{rule} → {next_state}"]))

    print("No solution found.")


dfs()
```
