```python
def nQueensSolver(size):
    positions = [-1] * size  
    all_answers = []

    def check_valid(current_row, current_col):
        r = 0
        while r < current_row:
            if positions[r] == current_col:
                return False
            col_diff = abs(positions[r] - current_col)
            row_diff = current_row - r
            if col_diff == row_diff:
                return False
            r += 1
        return True

    def place_queen(row_no):
        if row_no >= size:
            all_answers.append(positions.copy())
            return

        for column_no in range(size):
            if check_valid(row_no, column_no):
                positions[row_no] = column_no
                place_queen(row_no + 1)

        positions[row_no] = -1

    place_queen(0)
    return all_answers

num = int(input("Enter number of queens: "))
result_set = nQueensSolver(num)

print("\nTotal possible solutions:", len(result_set))

if result_set:
    print("\nOne solution is:\n")
    first = result_set[0]
    for r in range(num):
        row_display = ""
        for c in range(num):
            if first[r] == c:
                row_display += "Q "
            else:
                row_display += ". "
        print(row_display)
```
